```python
import csv
with open('Dataset/amazon_jobs_dataset.csv','r',encoding="utf8") as file_obj:
    file_data=csv.reader(file_obj,skipinitialspace=True)
    file_list=list(file_data) 
count=0
count1=0
count2=0

for row in file_list[1:]:
   if row[2][:2] == 'IN':
       p=row[5]
       if "Bachelor" in p or  "BS" in p or "BA" in  p:
            if "Java" in p:
                count+=1
            if "C++" in p:
                count1+=1
            if "Python" in p:
                count2+=1
ans = max(count, max(count1, count2))
if ans == count:
     print('Java', ans)
elif ans == count1:
     print('C++', ans)
```

```python
import numpy as np
#x=np.array([1,2,3,4])
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color='black',linewidth='5',label="x^3")
plt.plot(x,y2,color='blue',linewidth='5',label="x^2")
plt.axis([0,10,0,200]) ###changing the label
plt.ylabel('y')
plt.xlabel('x')
plt.legend()
plt.title("metplotDemo")
plt.show()
```

```python
import numpy as np
#x=np.array([1,2,3,4])
x=np.arange(0,5,0.1)
y=x**3
y2=x**2
plt.plot(x,y,color='black',linewidth='5',label="x^3")
plt.plot(x,y2,color='blue',linewidth='5',label="x^2")
plt.ylabel('y')
plt.xlabel('x')
plt.grid()
plt.text(2,80,"First Graph",fontsize=12)
plt.legend()
plt.title("metplotDemo")
plt.show()
```

### Plot Chart

```python
import matplotlib.pyplot as plt
year=[2012,2013,2014,2015,2016,2017]
salary=[12,13,14,17,19,20]
population=[100,120,180,250,300,370]
plt.scatter(year,salary,s=population,c='r')
plt.title="Bubble Chart"
plt.show()
```

```python
#Alpha
plt.scatter(year,salary,s=population,c='r',alpha=0.1)
plt.title="Bubble Chart"
plt.show()
```

```python
import numpy as np
c=np.arange(len(year))
c[1]=c[0]
plt.scatter(year,salary,s=population,c=c,alpha=0.5)
plt.title="Bubble Chart"
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black')
plt.title="Bubble Chart"
plt.show()
```

```python
plt.scatter(year,salary,s=population,c=c,alpha=0.5,edgecolor='black')
plt.title="Bubble Chart"
plt.text(year[0]+0.1,salary[0]+0.1,population[0])
plt.text(year[4]+0.2,salary[4]+0.2,population[4])
plt.show()
```

### Pie Chart

```python
import matplotlib.pyplot as plt
labels=['A','B','C','D']
colors=['blue','purple','red','pink']
size=[3,4,6,2]
explode=[0.1,0.2,0,0]
plt.pie(size,colors=colors,explode=explode,labels=labels,autopct='.2f%%',counterclock=False,startangle=100)
plt.title("Split AMONG CLASSS")
plt.axis("equal")#to make the shape round
plt.show()
```

```python
import matplotlib.pyplot as plt
company=['HP','Dell','Lenovo','Asus','Apple','Acer']
sold=[20000,43000,15000,17000,22000,13000]
plt.pie(sold,labels=company,autopct='%.1f',counterclock=False)
plt.axis('equal')
plt.show()
print('HP',15.4)
print('Dell',33.2)
print('Lenovo',11.5)
print('Asus',13.2)
print('Apple',16.9)
print('Acer',10.0)
```
